```python
#проект_Волкова, Исаева, Стрижкова, Трофимова, Щипакова
#считываем базу данных
import pandas as pd
df_raw = pd.read_excel('Downloads/dataset.xlsx')
df_raw
```

```python
#функция для различения имен святых и простых женщин
def saint(x):
    word = str(x).lower()
    if "свят" in word:
        return True
    else:
        return False
    
df_raw['святая'] = df_raw['социальный статус'].apply(saint)
df_raw
```

```python
#функция для отделение монашек от мирянок
def nun(x):
    word = str(x).lower()
    if 'инок' in word:
        return True
    elif 'скимн' in word:
        return True
    elif 'монах' in word:
        return True
    elif 'игумен' in word:
        return True
    else:
        return False
df_raw['монахиня'] = df_raw['социальный статус'].apply(nun)
df_raw
```

```python
#создаем новый датафрейм, исключая из первоначального упоминания святых
df = df_raw[df_raw['святая'] == False]
df
```

```python
#считываем файл со списком христианских канонических имен
christ_names = []
with open('names', mode='r', encoding='UTF-8') as file:
    for line in file:
        name = line.strip()
        christ_names.append(name)
print(christ_names)
```

```python
#определяем тип имени: языческое или христианское (относится к списку христианских имен)
def type_name(x):
    names = x.strip().split(': ')
    n = True
    for i in names:
        if i in christ_names:
            n = False
    return(n)
df['языческое имя'] = df['Канон имя + форма'].apply(type_name)
df
```

```python
#создаем функцию для определения того, является ли форма гипокористикой
def hyp(x):
    vowels = ['а', 'о', 'и', 'ы', 'у', 'и', 'э']
    names = x.strip().split(': ')
    n = False
    for i in names:
        if (i[-2] == "к") and (i[-3] not in vowels):
            n = True
    return(n)
df['гипокористика'] = df['Канон имя + форма'].apply(hyp)
df
```

```python
#создаем функцию для определения века, к которому относится текст
def century(x):
    date = str(x)
    if date == 'кон. XIV – нач. XV вв.':
        return(14)
    elif date == '970':
        return(10)
    elif date == 'nan':
        return(0)
    else:
        beg = date[0:4]
        beg = int(beg) // 100
        return (beg + 1)
df['век'] = df['Датировка'].apply(century)
df
```
